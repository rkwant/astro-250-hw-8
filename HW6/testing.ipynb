```python
import urllib2, json, urllibimport Image, StringIO, ImageFilter, ImageOps
```

```python
imgopener = urllib2.build_opener()imgopener.addheaders = [('User-agent', 'Mozilla/5.0')]imgpage = imgopener.open('http://upload.wikimedia.org/wikipedia/commons/thumb/d/d0/Poi_circles.jpg/280px-Poi_circles.jpg')my_picture = imgpage.read()		img = Image.open(StringIO.StringIO(my_picture))filtered = img.filter(ImageFilter.EDGE_ENHANCE_MORE)imshow(filtered, origin='bottom')
```

```python
imshow(ImageOps.autocontrast(filtered), origin='bottom')
```
