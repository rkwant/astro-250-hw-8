Part A------

```python
def rjSamp(f, ref, n):    """    given a pdf f and a reference distribution ref (eg stats.cauchy()),     this function generates n samples using rejection sampling        output: tuple (results, m, percent), where results are an array of the generated samples,    m, the m value used in the rejection sampling, and percent, the percentage of samples accepted    from the reference distribution    """    uniform = stats.uniform()    count = 0 #number of samples accepted    tries = 0 #total number of numbers generated    M = 0 #m-value to use    results = [] #array of numbers to return    sampleSize = n #number of results requested        #first, find an appropriate valuye for m by evaluating both distributions     #at a number of points and finding the largest value of f(x)/ref.pdf(x)    x = np.linspace(-50, 50, 100000)    gx = ref.pdf(x)    fx = f(x)    M = (fx/gx).max()        #iterate until we've accepted enough numbers    while count < n:        #generate sampleSize number of samples from the reference distribution and the uniform        rn = ref.rvs(sampleSize)        u = uniform.rvs(sampleSize)                #select the numbers to accept        newRvs = rn[where(u < (f(rn)/(M*ref.pdf(rn))))]        numNewRvs = len(newRvs) #determine how many numbers we added        count += numNewRvs #add to count        tries += sampleSize #add to tries                #add the numbers to the results array        results = np.hstack((results, newRvs))                #if we haven't accepted enough samples, determine an appropriate number of samples        #to take based on the acceptance rate from the previous try        if n-count >0:            sampleSize = int((float(sampleSize)/numNewRvs)*1.05*(n-count))        return (results[0:n], M, float(count)/tries)    
```

Part B------

```python
#define the laplace distributiondef laplace(x):    return (1./2)*exp(-abs(x))#define the CDF of the laplace distributiondef laplaceCDF(x):    if x>0:        return (1./2)*exp(x)    return (1./2)*(1-exp(-x))#define a vectorized cdf for later usedef lpCDF(x):    x = x.copy()    lt = np.where(x<0.)    gt = np.where(x>0.)    x[lt] = (1./2)*exp(x[lt])    x[gt] = 1.-((1./2)*exp(-x[gt]))    return x
```

```python
#generate 1000 samples rvs,m,percent=rjSamp(laplace, stats.cauchy(), 1000)print "The value of m was",m,"and the percent of samples accepted was",percent
```
