# PyAudioRecord from the computer's microphone.

```python
import pyaudiochunk = 1024FORMAT = pyaudio.paInt32CHANNELS = 1RATE = 44100RECORD_SECONDS = 5WAVE_OUTPUT_FILENAME = "output.wav"p = pyaudio.PyAudio()stream = p.open(format = FORMAT,    channels = CHANNELS,    rate = RATE,    input = True,    frames_per_buffer = chunk)all = []for i in range(0, RATE / chunk * RECORD_SECONDS):    data = stream.read(chunk)    all.append(data)print "* done recording"stream.close()p.terminate()
```

Write out the data to a file.

```python
import wavedata = "".join(all)wf = wave.open(WAVE_OUTPUT_FILENAME, "wb")wf.setnchannels(CHANNELS)wf.setsampwidth(p.get_sample_size(FORMAT))wf.setframerate(RATE)wf.writeframes(data)wf.close()
```
