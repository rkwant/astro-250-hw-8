```python
import SimpleXMLRPCServerfrom skimage import data, io, filter, color, img_as_float, transform, graph, img_as_int, img_as_ubyteio.use_plugin('matplotlib')io.use_plugin('pil', 'imread')import time
```

```python
class ImageManipulation:    '''Class to manipulate images over a server'''        '''    helper function that saves the input and output images from the function    that it wraps    '''    def introspect(f):        def wrapper(*arg,**kwarg):            t = time.strftime("%d_%b_%Y_%H-%M-%S", time.gmtime())            io.imsave('server_images/'+f.__name__+'_'+t+'_input.jpg', arg[1])            out = f(*arg,**kwarg)            io.imsave('server_images/'+f.__name__+'_'+t+'_output.jpg', out)            return out        return wrapper        @introspect    def channels(self, im):        """        input: takes a list of nested lists representing an image (nested lists can be generated by calling tolist() on a numpy array)        Output: a list of lists representing the image with the red and green channels reversed.  If the image        is grayscale, then the intensities are reversed (black becomes white, etc)        """        npa = array(im)        if len(npa.shape)>2:            one = npa[:,:,0].copy()            two = npa[:,:,1].copy()            npa[:,:,0] = two            npa[:,:,1] = one            return npa.tolist()        else:            inverse = npa.max()-npa            return inverse.tolist()                    @introspect    def upsideDown(self, im):        """        input: takes a list of nested lists representing an image (nested lists can be generated by calling tolist() on a numpy array)        Output: a list of lists representing the image turned upside down        """        npa = array(im)        return npa[::-1].tolist()        @introspect    def hSlice(self, im):        """        input: takes a list of nested lists representing an image (nested lists can be generated by calling tolist() on a numpy array)        Output: a list of lists representing the image with the right and left halves of the image transposed        """                npa = array(im)                if len(npa.shape)>2:            wd = len(npa[0,:,0])            return hstack((npa[:,wd/2:,:], npa[:,:wd/2,:])).tolist()        else:            wd = len(npa[0,:])            return hstack((npa[:,wd/2:], npa[:,:wd/2])).tolist()
```

```python
#start the serverhost, port = "", 5022server = SimpleXMLRPCServer.SimpleXMLRPCServer((host, port), allow_none=True)server.register_instance(ImageManipulation())server.register_multicall_functions()server.register_introspection_functions()print "XMLRPC Server is starting at:", host, port
```
