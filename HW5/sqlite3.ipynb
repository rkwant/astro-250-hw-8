# sqlite3 interface through PythonFirst, we show simple database creation.

```python
import sqlite3connection = sqlite3.connect("tmp/example.db")cursor = connection.cursor()sql_cmd = """CREATE TABLE dan_aykroyd (id INTEGER PRIMARY KEY AUTOINCREMENT,    skit_title TEXT, air_date DATE, season INT, ep INT, role TEXT)"""cursor.execute(sql_cmd)
```

Then, we insert some data.

```python
skit_data = [    ("Trojan Horse Home Security", "10/11/75", 1, 1, "Kenny Vorstrather"),     ("E. Buzz Miller's Animal Kingdom", "2/25/78", 3, 12, "E. Buzz Miller"),     ("The Coneheads at Home", "10/21/78", 4, 3, "Beldar Conehead")]for role in skit_data:    sql_cmd = ("INSERT INTO dan_aykroyd (skit_title, air_date, season, " +             "ep, role) VALUES " + str(role))    cursor.execute(sql_cmd)
```

Lastly, we run a simple query on the database we just made.

```python
sql_cmd = "SELECT * FROM dan_aykroyd WHERE air_date > '10/12/75'"cursor.execute(sql_cmd)db_info = cursor.fetchall()for entry in db_info: print entryconnection.commit()connection.close()
```

Create another table, enter data, run simple join on columns using both tables.

```python
connection = sqlite3.connect("tmp/example.db")cursor = connection.cursor()sql_cmd = """CREATE TABLE jane_curtin (id INTEGER PRIMARY KEY AUTOINCREMENT,    skit_title TEXT, air_date DATE, season INT, ep INT, role TEXT)"""cursor.execute(sql_cmd)skit_data = [    ("Teen Talk", "7/24/76", 1, 23, "Jane"),     ("The Snakehandling O'Sheas", "9/25/76", 2, 2, "Jane O'Shea"),     ("The Coneheads at Home", "10/21/78", 4, 3, "Prymaat Conehead")]for role in skit_data:    sql_cmd = ("INSERT INTO jane_curtin (skit_title, air_date, season, " +             "ep, role) VALUES " + str(role))    cursor.execute(sql_cmd)# Our query is to find which skits featured both Dan Aykroyd and Jane Curtin.sql_cmd = """SELECT dan_aykroyd.skit_title, dan_aykroyd.air_date,     dan_aykroyd.season, dan_aykroyd.ep, dan_aykroyd.role, jane_curtin.role    FROM dan_aykroyd, jane_curtin WHERE     dan_aykroyd.skit_title = jane_curtin.skit_title AND     dan_aykroyd.air_date = jane_curtin.air_date"""cursor.execute(sql_cmd)db_info = cursor.fetchall()for entry in db_info: print entry
```

We can also run a left join.

```python
sql_cmd = """SELECT dan_aykroyd.skit_title, dan_aykroyd.air_date,     dan_aykroyd.season, dan_aykroyd.ep, dan_aykroyd.role, jane_curtin.role    FROM dan_aykroyd LEFT JOIN jane_curtin ON     dan_aykroyd.skit_title = jane_curtin.skit_title AND     dan_aykroyd.air_date = jane_curtin.air_date"""cursor.execute(sql_cmd)db_info = cursor.fetchall()for entry in db_info: print entryconnection.commit()connection.close()
```

# Example with PlottingEmphasis on data retrieval

```python
import urllib2, sqlite3, datetimefrom numpy import loadtxt, array%pylab inlinedef get_stock_data(stock_symbol):    response = urllib2.urlopen("http://ichart.finance.yahoo.com/table.csv?s=" +         stock_symbol + "&d=9&e=3&f=2012&g=d&a=1&b=1&c=2000&ignore=.csv")    data = response.read()    response.close()    output = file(stock_symbol + ".txt", "w")    output.write(data)    output.close()    return loadtxt(stock_symbol + ".txt", skiprows=1, delimiter=",",                    dtype='|S10, float, float, float, float, float, float')
```

Here we create a table, download the stock data, and populate the table with the data.

```python
connection = sqlite3.connect(":memory:")cursor = connection.cursor()sql_cmd = """CREATE TABLE stocks (id INTEGER PRIMARY KEY AUTOINCREMENT,    symbol TEXT, day INT, open FLOAT, high FLOAT, low FLOAT, close FLOAT,    volume FLOAT, adj_close FLOAT)"""cursor.execute(sql_cmd)stock_symbols = ["VZ", "AAPL", "FORD", "MSFT", "BA"]ref_date = datetime.date(2000, 1, 1)for stock_symbol in stock_symbols:    data = get_stock_data(stock_symbol)    for row in data:        year = int(row[0].split("-")[0])        month = int(row[0].split("-")[1])        day = int(row[0].split("-")[2])        stock_date = datetime.date(year, month, day)        row[0] = (stock_date - ref_date).days        sql_cmd = ("INSERT INTO stocks (symbol, day, open, high, low, close, " +            "volume, adj_close) VALUES " +             str(row).replace("(", "('" + stock_symbol + "', "))        cursor.execute(sql_cmd)    connection.commit()
```

We can plot from the table easily, using the query to select out only the data we want to visualize.

```python
figsize(12,10)for ss in stock_symbols:    sql_cmd = "SELECT day, close FROM stocks WHERE symbol == '" + ss + "'"    cursor.execute(sql_cmd)    db_info = array(cursor.fetchall())    plot(db_info[:, 0], db_info[:, 1], label=ss)connection.close()legend(loc=2)xlabel("Days since 1/1/2000")ylabel("Stock Value")suptitle("Stock Performance")
```
